```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

plt.rcParams['font.family'] = 'Microsoft JhengHei'  # 設置字體
plt.rcParams['font.size'] = 12 # 設置字體大小
```

```python
# 很多人在买一本书的时候，都比较喜欢货比三家，例如《python数据分析实战》在亚马逊、当当网、中国图书网、京东和天猫的
#最低价格分别为38.8,40.2,48.4,39.9,32.34。针对这个数据，我们也可以通过条形图来完成，这里使用水平条形图来显示：
store=['亚马逊','当当网','中国图书网','京东','天猫']
price=[38.8,40.2,48.4,39.9,32.34]
# 水平条形图
plt.figure(figsize=(6,4),dpi=100)  # figsize (寬,高)
# 画折线图
plt.barh(store,price,color='green',alpha=0.8)
for x,y in enumerate(price):
    plt.text(y+2,x,y,ha='center')
plt.show()
```
